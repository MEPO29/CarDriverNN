```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def plot_checkpoints(image_path, coordinates):
    """
    Carga una imagen y grafica puntos en las coordenadas especificadas.

    Parameters:
    - image_path (str): Ruta de la imagen de fondo.
    - coordinates (list of tuples): Lista de tuplas (x, y) que representan las coordenadas de los checkpoints.
    """
    # Cargar la imagen
    img = mpimg.imread(image_path)

    # Crear la figura
    plt.figure(figsize=(10, 6))
    plt.imshow(img)

    # Separar las coordenadas en listas de x e y
    x_coords = [x for x, y in coordinates]
    y_coords = [y for x, y in coordinates]

    # Graficar los puntos
    plt.scatter(x_coords, y_coords, c='red', s=100, marker='o', label='Checkpoints')  # Puntos rojos con tamaño y forma especificados

    # Etiquetas y título
    plt.title("Checkpoints en la Pista")
    plt.xlabel("Eje X (pixeles)")
    plt.ylabel("Eje Y (pixeles)")

    # Mostrar cuadrícula y leyenda
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()

    # Mostrar la imagen y los puntos
    plt.axis('on')  # Mostrar los ejes
    plt.show()

#coordenadas=[(120, 480), (200, 400), (300, 300), (400, 200), (500, 100)]
coordenadas=[
        (100, 500),  # Checkpoint 1 para bg7
        (350, 300),
        (520,580),# Checkpoint 2
        (800, 250),
        (1400,100),
        (1200,550),# Checkpoint 3
        (500, 780)   # Checkpoint 5
    ]
plot_checkpoints("bg7.png",coordenadas)
```
